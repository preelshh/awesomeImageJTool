```python
import pandas as pd
```

```python
df_A = pd.read_csv(r"C:\Users\priya\results\df_A.csv")
df_B = pd.read_csv(r"C:\Users\priya\results\df_B.csv")
df_C = pd.read_csv(r"C:\Users\priya\results\df_C.csv")
df_D = pd.read_csv(r"C:\Users\priya\results\df_D.csv")
df_A['particle_type_human'] = 'A'
df_B['particle_type_human'] = 'B'
df_C['particle_type_human'] = 'C'
df_D['particle_type_human'] = 'D'
df_all = pd.concat([df_A, df_B, df_C, df_D], ignore_index=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df_all,
    x='Area',
    y='Perimeter',
    hue='particle_type_human',
    palette='Set2',
    s=80,
    alpha=0.7
)

plt.title('Area vs. Perimeter by Morphological Type')
plt.xlabel('Area')
plt.ylabel('Perimeter')
plt.legend(title='Logic Class')`
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Compute mean of each group
mean_points = df_all.groupby('particle_type_human')[['Solidity', 'Convexity']]]

# Plot
plt.figure(figsize=(7, 6))
for _, row in mean_points.iterrows():
    plt.scatter(row['Solidity'], row['Convexity'], label=row['particle_type_human'], s=150)

plt.title('Mean Convexity vs. Solidity by Morphological Type')
plt.xlabel('Solidity')
plt.ylabel('Convexity')
plt.legend(title='Perimeter Type')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Compute mean of each group
mean_points = df_all.groupby('particle_type_human')[['Area', 'Perimeter']].mean().reset_index()

# Plot
plt.figure(figsize=(7, 6))
for _, row in mean_points.iterrows():
    plt.scatter(row['Area'], row['Perimeter'], label=row['particle_type_human'], s=150)

plt.title('Mean Area vs. Perimeter by Morphological Type')
plt.xlabel('Area')
plt.ylabel('Perimeter')
plt.legend(title='Particle Type')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Compute mean of each group
mean_points = df_all.groupby('particle_type_human')[['Tortuosity', 'GeodesicElongation']].mean().reset_index()

# Plot
plt.figure(figsize=(7, 6))
for _, row in mean_points.iterrows():
    plt.scatter(row['Tortuosity'], row['GeodesicElongation'], label=row['particle_type_human'], s=150)

plt.title('Mean Tortuosity vs. Geodesic Elongation by Morphological Type')
plt.xlabel('Tortuosity')
plt.ylabel('Geodesic Elongation')
plt.legend(title='Particle Type')
plt.grid(True)
plt.tight_layout()
plt.show()
```
